```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent = dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'absolute_error', splitter = 'best')
regressor = regressor.fit(x_train, y_train)

#1. criterion = 'squared_error', splitter = 'best'    -> 0.9057295380760667
#2. criterion = 'squared_error', splitter = 'random'  -> 0.9422695581792044
#3. criterion = 'friedman_mse', splitter = 'best'     -> 0.9293400506580218
#4. criterion = 'friedman_mse', splitter = 'random'   -> 0.8978348862113252
#5. criterion = 'absolute_error', splitter = 'best'   -> 0.947098522115071   << gives best r square value
#6. criterion = 'absolute_error', splitter = 'random' -> 0.9265306845303183
#7. criterion = 'poisson', splitter = 'best'          -> 0.9466973756666384
#8. criterion = 'poisson', splitter = 'random'        -> 0.9091284786302676
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```
