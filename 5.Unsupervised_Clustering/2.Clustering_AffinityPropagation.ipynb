```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x = dataset.iloc[:, 3:5].values
x
```

```python
from sklearn.cluster import AffinityPropagation

affPro = AffinityPropagation(random_state = 5)
yAff   = affPro.fit_predict(x)

yAff
```

```python
dataset['Cluster_Group'] = yAff
dataset
```

```python
affPro.cluster_centers_
```

```python
import seaborn as sb

facet = sb.lmplot(data = dataset, x = dataset.columns[3], y = dataset.columns[4], hue = dataset.columns[5], 
                  fit_reg = False, legend = True, legend_out = True)
```
