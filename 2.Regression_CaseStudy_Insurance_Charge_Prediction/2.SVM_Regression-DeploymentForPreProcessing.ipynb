```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

#Pre-processing the output values
scy = StandardScaler()
y_train = scy.fit_transform(y_train)
y_test  = scy.transform(y_test)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVR

regressor = SVR(kernel = "rbf", C=5000)


#1. kernel = linear, C=10     ->  0.46246841423396834
#2. kernel = linear, C=100    ->  0.6288792857320361
#3. kernel = linear, C=1000   ->  0.7649311738597486
#4. kernel = linear, C=2000   ->  0.7440418308107889
#5. kernel = linear, C=3000   ->  0.7414236599248987
#6. kernel = linear, C=5000   ->  0.7414179301703103

#7. kernel = rbf, C=10        -> -0.0322732939067103
#8. kernel = rbf, C=100       ->  0.3200317832050832
#9. kernel = rbf, C=1000      ->  0.8102064874808204
#10. kernel = rbf, C=2000     ->  0.8547766422240716
#11. kernel = rbf, C=3000     ->  0.8663393963090398
#12. kernel = rbf, C=5000     ->  0.874777817423661           <<<<<<<<

#13. kernel = poly, C=10      ->  0.038716222760231456
#14. kernel = poly, C=100     ->  0.6179569624059796
#15. kernel = poly, C=100     ->  0.8566487675946555
#16. kernel = poly, C=2000    ->  0.8605579276143769
#17. kernel = poly, C=3000    ->  0.8598930084494327
#18. kernel = poly, C=5000    ->  0.8595656374126361

#19. kernel = sigmoid, C=10   ->  0.03930714378274347
#20. kernel = sigmoid, C=100  ->  0.5276103546510407
#21. kernel = sigmoid, C=1000 ->  0.2874706948697727
#22. kernel = sigmoid, C=2000 ->  -0.5939509731283508
#23. kernel = sigmoid, C=3000 ->  r score is deteriorating, no need to test
#24. kernel = sigmoid, C=5000 ->  r score is deteriorating, no need to test
```

```python
regressor.fit(x_train, y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```
