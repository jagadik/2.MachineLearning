```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion = 'poisson', n_estimators = 100, random_state = 0)
regressor = regressor.fit(x_train, y_train)

#1. criterion = 'squared_error', n_estimators = 50    -> 0.8495860472309916
#2. criterion = 'squared_error', n_estimators = 100   -> 0.8537074492312178
#3. criterion = 'absolute_error', n_estimators = 50   -> 0.8533104199010396
#4. criterion = 'absolute_error', n_estimators = 100  -> 0.8522171666048011
#5. criterion = 'friedman_mse', n_estimators = 50     -> 0.8498058213339406
#6. criterion = 'friedman_mse', n_estimators = 100    -> 0.8540807721486975    <<<<<<
#7. criterion = 'poisson', n_estimators = 50          -> 0.8491113222296434
#8. criterion = 'poisson', n_estimators = 100         -> 0.8526481325996583
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```
