```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion' : ['poisson', 'friedman_mse', 'absolute_error', 'squared_error'], 'max_features' : ['auto', 'sqrt', 'log2'], 'n_estimators' : [10, 100]}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose = 3, n_jobs = -1)

#removed test,train split and also removed Scaler. Passing the whole data to fit method
grid.fit(independent, dependent)
```

```python
result = grid.cv_results_
print("The R_Score value for best parameter {}:", format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(result)
table
```

```python
prediction_result = grid.predict([[43, 56, 2, 1, 1]])
prediction_result
```
