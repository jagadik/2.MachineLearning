```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'poisson', splitter = 'random')
regressor = regressor.fit(x_train, y_train)

#1. criterion = 'squared_error', splitter = 'best'    -> 0.6935643620288746
#2. criterion = 'squared_error', splitter = 'random'  -> 0.6786146385639003
#3. criterion = 'friedman_mse', splitter = 'best'     -> 0.6851078940595519
#4. criterion = 'friedman_mse', splitter = 'random'   -> 0.6677158531404498
#5. criterion = 'absolute_error', splitter = 'best'   -> 0.6602322905539526
#6. criterion = 'absolute_error', splitter = 'random' -> 0.7455071283652015   <<<<<<
#7. criterion = 'poisson', splitter = 'best'          -> 0.7201600296042935
#8. criterion = 'poisson', splitter = 'random'        -> 0.7013998594969812
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```
